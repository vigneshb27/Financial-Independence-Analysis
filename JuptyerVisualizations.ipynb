```python
import pandas as pd
import matplotlib.pyplot as plt

plt.hist(dataset['age_when_financially_literate'], bins=10)
plt.xlabel('Age when Financially Literate')
plt.ylabel('Frequency')
plt.title('Histogram of Age when Financially Literate')
plt.show()
```

```python
import seaborn as sns
plt.scatter(dataset['age_when_financially_literate'], dataset['frugality'])
plt.xlabel('Age when Financially Literate')
plt.ylabel('Frugality')
plt.title('Scatter Plot: Age when Financially Literate vs. Frugality')
plt.show()
```

```python
sns.boxplot(x=dataset['education'], y=dataset['age_when_financially_literate'])
plt.xlabel('Education')
plt.ylabel('Age when Financially Literate')
plt.title('Box Plot: Education vs. Age when Financially Literate')
plt.xticks(rotation=45)
plt.show()
```

```python
import numpy as np

numeric_columns = ['age', 'gross_annual_wage', 'total_portfolio']
numeric_data = dataset[numeric_columns]
corr_matrix = numeric_data.corr()
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))

sns.heatmap(corr_matrix, mask=mask, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
numeric_columns = ['age', 'gross_annual_wage', 'total_portfolio']
numeric_data = dataset[numeric_columns]

sns.pairplot(numeric_data)
plt.title('Pair Plot of Numeric Variables')
plt.show()
```

```python
plt.boxplot(dataset['age'])
plt.xlabel('Age')
plt.title('Box Plot of Age')
plt.show()
```

```python
plt.pie(dataset['education'].value_counts(), labels=dataset['education'].unique(), autopct='%1.1f%%')
plt.title('Pie Chart of Education')
plt.axis('equal')
plt.show()
```

```python
sns.barplot(x=dataset['education'], y=dataset['age'], hue=dataset['gender'])
plt.xlabel('Education')
plt.ylabel('Age')
plt.title('Grouped Bar Plot: Education vs. Age by Gender')
plt.xticks(rotation=45)
plt.show()
```

```python
numeric_columns = ['age', 'gross_annual_wage', 'total_portfolio']
numeric_data = dataset[numeric_columns]

sns.pairplot(numeric_data)
plt.title('Scatter Plot Matrix of Numeric Variables')
plt.show()
```

```python
cross_tab = pd.crosstab(index=dataset['education'], columns=[dataset['job'], dataset['housing_situation']])
cross_tab.plot(kind='bar', stacked=True)
plt.xlabel('Education')
plt.ylabel('Count')
plt.title('Stacked Bar Plot: Education vs. Job and Housing Situation')
plt.xticks(rotation=45)
plt.legend(title=['Job', 'Housing Situation'])
plt.show()
```

```python
plt.scatter(dataset['age'], dataset['gross_annual_wage'], c=dataset['total_portfolio'], cmap='viridis')
plt.xlabel('Age')
plt.ylabel('Gross Annual Wage')
plt.title('Scatter Plot: Age vs. Gross Annual Wage with Total Portfolio Color Mapping')
plt.colorbar(label='Total Portfolio')
plt.show()
```

```python
sns.kdeplot(dataset['gross_annual_wage'], shade=True)
plt.xlabel('Gross Annual Wage')
plt.ylabel('Density')
plt.title('Kernel Density Estimation Plot of Gross Annual Wage')
plt.show()
```

```python
sns.violinplot(x=dataset['job'], y=dataset['age_when_financially_literate'])
plt.xlabel('Job')
plt.ylabel('Age when Financially Literate')
plt.title('Violin Plot: Job vs. Age when Financially Literate')
plt.xticks(rotation=45)
plt.show()
```

```python
cross_tab = pd.crosstab(index=dataset['education'], columns=dataset['job'])
cross_tab.plot(kind='bar')
plt.xlabel('Education')
plt.ylabel('Count')
plt.title('Clustered Bar Plot: Education vs. Job')
plt.xticks(rotation=45)
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(dataset['age'], dataset['gross_annual_wage'], dataset['total_portfolio'])
ax.set_xlabel('Age')
ax.set_ylabel('Gross Annual Wage')
ax.set_zlabel('Total Portfolio')
plt.title('3D Scatter Plot: Age, Gross Annual Wage, Total Portfolio')
plt.show()
```
